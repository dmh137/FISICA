## 1.1 VIBRACION LIBRE DE UN SISTEMA NO AMORTIGUADO

PARA UNA MASA QUE SE DESPLAZA UNA DISTANCIA X CUANDO UNA FUERZA RESULTANTE F ACTÚA EN ELLA EN LA MISMA DIRECCIÓN

$$F=M\times A$$

PARA UN CUERPO RÍGICO SOMETIDO A MOVIMIENTO DE ROTACIÓN 

$$\overrightarrow{M}(t)=J\ddot{\theta }$$

Princripio de la conservación: Un sistema es conservativo si se dice que no pierde energia debido a la fricción o a miembros no elásticos  que disipan la energía.

La energía cinetica T se almacena en la masa por efecto de la velocidad y la energía potencial U se almacena en el resorte a causa de la deformacion elastica. Lo cual lleva a:

$$T+U=constante$$

$$T=\frac{1}{2}m\dot{x}^{2}$$

$$U=\frac{1}{2}kx^{2}$$

La solucion de la ecuacion se obtiene de manera general

$$x(t)=C_{1}e^{w_{n}t}+C_{2}e^{-w_{n}t}$$

O tambien:

$$x(t)=A_{1}cos w_{n}t+A_{2}sen w_{n}t$$

Tambien se puede expresar como 

$$x=x_{m}sen (\omega_{n}t+\varphi)$$

Los valores de $A_{1}$ y $A_{2}$ se obtienen de las condiciones dadas

Otras formulas

Frecuencia natural para no amortiguado:
$$\omega_{n}=\sqrt{\frac{k}{m}}$$  

Periodo:
$$\tau=\frac{2\pi}{\omega_{n}}$$

Frecuencia:
$$f=\frac{1}{\tau}$$
